```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
```

```python
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990) & 
                        (netflix_df['release_year'] <= 1999) & 
                        (netflix_df['type'] == 'Movie')]
print(movies_90s['duration'].dtype)
#.mode(): This returns the most frequent value(s) in the column. The [0] index is used to get the first mode in case there are multiple.
duration = movies_90s['duration'].mode()[0]
print(duration)
'''we can work with a for loop :
# Filter for movies released in the 1990s
movies_90s = netflix_df[(netflix_df['release_year'] >= 1990) & 
                        (netflix_df['release_year'] <= 1999) & 
                        (netflix_df['type'] == 'Movie')]

# Initialize an empty dictionary to count the occurrences of each duration
duration_count = {}

# Iterate over each row in the filtered DataFrame
for lab, row in movies_90s.iterrows():
    duration = row['duration']
    
    # Update the count for each duration in the dictionary
    if duration in duration_count:
        duration_count[duration] += 1
    else:
        duration_count[duration] = 1

# Find the duration with the maximum count (most frequent)
most_frequent_duration = max(duration_count, key=duration_count.get)

print(most_frequent_duration)
'''

```
